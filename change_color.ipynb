```python
import numpy as np
import cv2
import matplotlib.pylab as plt
```

```python
def show_img(img):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(img)
    plt.show()
    
```

```python
img = cv2.imread("img2/FILE1139.png")
img_ori = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
show_img(img_ori)
```
